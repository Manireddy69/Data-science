```python
from sklearn.feature_extraction import FeatureHasher
vgsales_df = pd.read_csv('vgsales.csv')
vgsales_df.head()
```

```python
vgsales_df.info()
```

```python
#To select diretly dtype = object
vgsales_df.select_dtypes(include= 'object').head()
```

```python
u_generes= vgsales_df['Genre'].unique()
# len(vgsales_df['Genre'].unique())
u_generes
```

```python
fh = FeatureHasher(n_features= 6, input_type='string')
hashed_features = fh.fit_transform(vgsales_df['Genre'].apply(lambda x: [x]))
hashed_features = hashed_features.toarray()
new_game_df = pd.concat([vgsales_df[['Name','Genre']], pd.DataFrame(hashed_features)],axis= 1)
new_game_df.head()
```
