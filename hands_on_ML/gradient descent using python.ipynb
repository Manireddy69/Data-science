```python
import numpy as np
import matplotlib.pyplot as plt
# Define the function and its gradient
def f(x):
    return x**2  #f(x) = x

def gradient(x):
    return 2 * x #partial derivative of f(x) with respect to x

# Initialize variables
x = 10
learning_rate = 0.1
steps = 20
x_values = [x]
f_values = [f(x)]

# Run gradient descent and store values
for i in range(steps):
    x = x - learning_rate * gradient(x)
    x_values.append(x)
    f_values.append(f(x))

# -----------------------------
#  2D Visualization
# -----------------------------
x_plot = np.linspace(-10, 10, 140)  # Range for x-axis
y_plot = f(x_plot)

plt.figure(figsize=(10, 5))
plt.plot(x_plot, y_plot, label='f(x) = x')
plt.scatter(x_values, f_values, color='red', label='Gradient Descent Steps')
plt.plot(x_values, f_values, linestyle='--', color='gray')
plt.title("2D Gradient Descent Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
