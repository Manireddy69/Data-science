```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
list(iris.keys())
```

```python
X = iris['data'][:, 3:]
y = (iris['target'] ==2).astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
iris.feature_names
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)
plt.plot(X_new, y_prob[:, 1],'g-', label = "Iris_virgina")
plt.plot(X_new, y_prob[:, 0], 'b--', label = "Not Iris-virgina")
plt.legend()
```

```python
y_prob[:,0].shape
```

```python
log_reg.predict([[1.7],[0.5]]) #this is petal length and width
```

```python
#softmax regrsssion(used for multiclasses)
print(iris.feature_names)
X = iris['data'][:,(2,3)] #we take only petal length & width
y = iris['target']
```

```python
soft_max = LogisticRegression(multi_class ='multinomial', solver = 'lbfgs', C = 10)
soft_max.fit(X,y)
#{‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘newton-cholesky’, ‘sag’, ‘saga’}, default=’lbfgs’
```

```python
soft_max.predict([[5,2]]) #[petal lenght, petal width]
#array([2]) means it is "class 2"
```

```python
soft_max.predict_proba([[5,2]])
```
