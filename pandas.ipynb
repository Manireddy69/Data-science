```python
import os
import pandas as pd

# List files in the dataset directory
file_names = os.listdir(path)
print("Files in the 'youtube-new' dataset directory:")
for file_name in file_names:
    print(file_name)

# Assuming one of the main files is 'USvideos.csv', let's load it
# You can change 'USvideos.csv' to any other CSV file name you see above
# if you want to load a different country's data.
csv_file_path = os.path.join(path, 'USvideos.csv')
df_youtube = pd.read_csv(csv_file_path)

# Display the first 5 rows of the DataFrame
display(df_youtube.head())
```
